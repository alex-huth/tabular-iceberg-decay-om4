```python
def pre_processing(r1,number,tot):
    print('pre_processing',number,'of ',tot)
    m2_to_km2 = 1.e-6
    r1=r1.where(r1['lat']>0,drop=True)
    r1['area'] = m2_to_km2*r1['mass']/(rhoi*r1['thickness']); r1=r1.where(r1['area']>=0.5,drop=True)
    drop=['id_cnt','id_ij']
    r1['id'] = r1['id_cnt']*2**32+r1['id_ij']; r1=r1.drop(drop)
    return r1
```

```python
#centers the range of longitude on midlon
def center_lon(ds,midlon2,is_xr):
    k=1
    minlon2=midlon2-180
    maxlon2=midlon2+180
    while (k == 1):
        k=0
        if is_xr:
            #xarray
            if (np.any(ds['lon']>maxlon2)):
                ds['lon']=xr.where(ds['lon']>maxlon2,ds['lon']-360,ds['lon']); k=1
            if (np.any(ds['lon']<minlon2)):
                ds['lon']=xr.where(ds['lon']<minlon2,ds['lon']+360,ds['lon']); k=1
        else:
            #pandas
            if (np.any(ds['lon']>maxlon2)):
                ds.lon=ds.lon.mask(ds.lon>maxlon2,ds.lon-360); k=1
            if (np.any(ds['lon']<minlon2)):
                ds.lon=ds.lon.mask(ds.lon<minlon2,ds.lon+360); k=1
    return ds

def redo_modeled_array(r1,minlat,maxlat):
    print('redo_modeled_array...')
    r1=r1.where(r1['lat']>0,drop=True)
    r1=r1.where(r1['lat']<=maxlat,drop=True)
    r1=r1.where(r1['lat']>=minlat,drop=True)
    #r1=r1.where(r1['area']>=0.5,drop=True)
    r1['time']=r1['year']-1990+r1['day']/370.
    bergs=list(set(r1['id'].values.astype(int))); bergs=np.sort(np.array(bergs))
    r1b=r1.drop(['year','day']); r1b=r1b.sortby(['id','time']); r1b['age']=r1b['time'].copy()*0
    r1bidvals=r1b.id.values; leng=len(r1bidvals); f_index=np.zeros(len(bergs)); cid=-1; k=0

    for i,val in enumerate(r1bidvals):
        if cid!=val:
            cid=val; f_index[k]=i; k+=1
    f_index=f_index.astype(int); l_index=f_index[1:len(f_index)]; l_index=np.append(l_index,leng)

    for i in range(len(f_index)):
        r1b['age'][f_index[i]:l_index[i]]=r1b['time'][f_index[i]]
    r1b['age']=r1b['time']-r1b['age']
    print('done')
    return r1b
```

# Get number of unique and total bergs, and average berg area and age, within 100 km^2 of a grid node

-work in slices of longitude, adjusting so that lat/lon for both bergs and mesh is centered at some point, and adjusted so that there is no wrap-around.
-In OM4p5, the northern hemisphere grid uses an irregular bipolar cap (Murray, 1966) north of ~65 degrees N. Therefore, the fl_traj_to_mesh routines in the following cell are slightly different than those used in the S. Hemisphere (see fl_traj_to_mesh_SH.ipynb).

```python
def haversine_dist(lon1,lat1,lon2,lat2):
    R=6360000

    p1 = lat1 * np.pi/180
    p2 = lat2 * np.pi/180
    dp = (lat2-lat1) * np.pi/180
    dm = (lon2-lon1) * np.pi/180

    a = np.sin(0.5*dp)**2 + np.cos(p1) * np.cos(p2) * np.sin(0.5*dm)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))

    d = R * c
    return d
    
def fl_traj_to_mesh_2(r_mesh,rp,kernel,maxlatin,minlatin,donecode):
    print('starting fl_traj_to_mesh...')
    start=time.time()

    #constants
    latcut=64.9
    target_dist_km=kernel
    Rearth=6360000.; mult=(np.pi/180.); mult2=(mult*Rearth)**2
    mult=0.5*mult
    
    #faster way (not using haversine dist). Gives indistinguishable result as haversine approach:
    #target_dist=(target_dist_km*1e3)**2
    #most accurate way (haversine dist):
    target_dist=target_dist_km*1e3
    
    latm=min(maxlatin+0.01,89.9999) #a small buffer
    lonm=0; lonm2=1; latm2=latm

    #faster way (not using haversine dist). Gives indistinguishable result as haversine approach:
    #min_gx=np.sqrt(mult2*(((lonm-lonm2)*np.cos((latm+latm2)*mult))**2))/1000
    #most accurate way (haversine):
    min_gx=haversine_dist(lonm,latm,lonm2,latm2)/1000
    
    p_lon_buffer=target_dist_km/min_gx + 0.1 #this buffer is used to extend the sector of particle longitude in the below code

    min_gy=np.sqrt(mult2)/1000
    p_lat_buffer=target_dist_km/min_gy + 0.1
        
    print('lon buffer',p_lon_buffer, 'lat buffer',p_lat_buffer)
    
    minlat_tp=minlatin-p_lat_buffer; maxlat_tp=maxlatin+p_lat_buffer
    
    if donecode==1:
        minlatin=minlat_tp
    
    modeled_ub_len=len(set(rp.id.values))
    modeled=redo_modeled_array(rp,minlat_tp,maxlat_tp)

    modeled['dist']=modeled['area'].copy()*0
    #modeled_ub_len=len(set(modeled.id.values))
    modeled_len=len(modeled.id)

    print('modeled: total number of unique bergs (all locs)',modeled_ub_len)
    print('modeled: total number of berg samples (current locs)',modeled_len)
    print('modeled: max lat',modeled.lat.max())
    
    mesh_n=r_mesh.copy()
    mesh_n.drop(labels=['yTe','xTe'])
    print('minlat',minlatin,'maxlat',maxlatin)
    mesh_n=mesh_n.where(((mesh_n['lat']>=minlatin) & (mesh_n['lat']<=maxlatin)),drop=True)

    #min,mid,and max lons for grid and particle arrays, for splitting into 10 loops
    num_slices=41
    minlon_g=np.linspace(0,360,num_slices)[0:-1]
    seg=minlon_g[1]-minlon_g[0]
    maxlon_g=minlon_g+seg #for grid
    midlon=minlon_g+seg/2 #for grid and particles
    minlon_p=minlon_g-p_lon_buffer
    maxlon_p=maxlon_g+p_lon_buffer #for particles
    
    #print('midlons',midlon)
    print('starting iters...')
    #now add area and other stuff
    start=time.time()
    
    print('lon',modeled.lon)
    
    for iter in range(len(midlon)):
        curtime=time.time()
        if ((iter+1)%10==0):
            print('iter',iter+1,'of',len(midlon),'Time',curtime-start)

        #center longitude on midlon and extract smaller arrays
        mesh_n=center_lon(mesh_n,midlon[iter],is_xr=True)
        modeled=center_lon(modeled,midlon[iter],is_xr=True)
        modeled1=modeled.copy()
        modeled1=modeled1.where((modeled1['lon']>=minlon_p[iter]) & (modeled1['lon']<=maxlon_p[iter]),drop=True)
        
        if (len(modeled1.i)==0):
            mesh_n['done']=xr.where((mesh_n['lon']>=minlon_g[iter]) & (mesh_n['lon']<=maxlon_g[iter]),donecode,mesh_n['done'])
            continue

        for i in range(len(mesh_n['xT'])):
            
            if (maxlatin<=latcut):
                lonm=mesh_n['lon'][:,i].max().values
                if ((lonm<=minlon_g[iter]) | (lonm>maxlon_g[iter]) | (np.isnan(lonm))):
                    continue

            for j in range(len(mesh_n['yT'])):
                
                if (mesh_n['done'][j,i]>0):
                    continue
                
                if (maxlatin>latcut):
                    lonm=mesh_n['lon'][j,i].values
                    if ((lonm<=minlon_g[iter]) | (lonm>maxlon_g[iter]) | (np.isnan(lonm))):
                        continue
                        
                latm=mesh_n['lat'][j,i].values

                if ((np.isnan(latm)) or (latm>maxlatin)):
                    continue                

                maxlat=latm+p_lat_buffer; minlat=latm-p_lat_buffer
                
                modeled2=modeled1.copy()
                modeled2=modeled2.where(((modeled2['lat']<=maxlat) & (modeled2['lat']>=minlat)),drop=True)
                if (len(modeled2.i)==0):
                    continue
                
                #faster way (not using haversine dist). Gives indistinguishable result as haversine approach:
                #modeled2['dist']=mult2*(((lonm-modeled2['lon'])*np.cos(mult*(latm+modeled2['lat'])))**2+(latm-modeled2['lat'])**2)
                #most accurate way (haversine):
                modeled2['dist']=haversine_dist(lonm,latm,modeled2['lon'],modeled2['lat'])

                modeled2=modeled2.where(modeled2['dist']<target_dist,drop=True)                                 
                
                mesh_n['done'][j,i]=donecode
                
                if (len(modeled2.i)==0):
                    continue

                bset2=set(modeled2['id'].values.astype(int))
                mesh_n['u_bcount_modeled'][j,i]=len(bset2)

                avs=np.zeros((len(bset2),4))

                for kk,theid in enumerate(bset2):

                    modeledid=modeled2.where(modeled2['id']==theid,drop=True)

                    avs[kk,0]=modeledid['age'].mean()
                    avs[kk,1]=modeledid['area'].mean()
                    avs[kk,2]=modeledid['area'][modeledid['area']>=0.75].mean()
                    avs[kk,3]=modeledid['area'][modeledid['area']>=1.0].mean()

                with warnings.catch_warnings():
                    warnings.simplefilter("ignore", category=RuntimeWarning)
                    avs=np.nanmean(avs,0)
                
                avs=np.where(np.isnan(avs),0,avs)

                mesh_n['u_bage_modeled'][j,i]=avs[0]
                mesh_n['u_barea_modeled'][j,i]=avs[1]
                mesh_n['u_barea_modeled_overp75kms'][j,i]=avs[2]
                mesh_n['u_barea_modeled_over1kms'][j,i]=avs[3]                
                 
    mesh_n['u_bcount_modeled']=mesh_n['u_bcount_modeled']/modeled_ub_len

    varlist=['u_bcount_modeled','u_barea_modeled','u_barea_modeled_overp75kms','u_barea_modeled_over1kms','u_bage_modeled','done']
    for var in varlist:

        r_mesh['tempvar']=r_mesh['mask']
        r_mesh['tempvar']=mesh_n[var].where((mesh_n['lat']>=minlatin) & (mesh_n['lat']<=maxlatin))
        r_mesh[var]=xr.where((r_mesh['lat']>=minlatin) & (r_mesh['lat']<=maxlatin),r_mesh['tempvar'],r_mesh[var])
        r_mesh[var]=r_mesh[var].where(r_mesh['mask']>0)
        
    return r_mesh

def fl_traj_to_mesh(ms,r1x,kernel,maxlatin1,minlatin1):
    start=time.time()
        
    varlist=['u_bcount_modeled','u_barea_modeled','u_barea_modeled_overp75kms','u_barea_modeled_over1kms','u_bage_modeled','done']
    for var in varlist:
        ms[var]=ms['mask'].copy()
        ms[var][:,:]=0
        
    ms2=ms.copy()
    ms3=ms.copy()
    
    mm=64.8
    latcut=64.9
    
    if (maxlatin1<=latcut):
        print('run 1/1')
        ms3=fl_traj_to_mesh_2(ms2,r1x,kernel,maxlatin1,minlatin1,1)
    else:
        print('run 1/2')
        ms2=fl_traj_to_mesh_2(ms2,r1x,kernel,mm,minlatin1,1)
        print('run 2/2')
        ms3=fl_traj_to_mesh_2(ms2,r1x,kernel,maxlatin1,mm,2)
        ms4=xr.where(ms3['lat']<=64.8,ms2,ms3)
    tot_time=time.time()-start
    print('total time (min)',tot_time/60)
    print('')
    print('')
    return ms4
```

```python
def fixlatlabel_gl(gl):

    gl.x_inline=False
    gl.y_inline=False
    gl.xlabel_style = {'size': 15, 'color': 'gray'}
    gl.ylabel_style = {'size': 15, 'color': 'gray'}
    gl.rotate_labels = False
    plt.draw()
    return gl

def plotformatting(mp,ax):    
    dir_labels=True
    
    xticks=[-75,-50,-25,0]; yticks=[45,60]  
    proj_to_data = ccrs.PlateCarree()._as_mpl_transform(ax) - ax.transData
    bound_in_target = proj_to_data.transform_path(boundary_path)
    ax.set_extent(extent2, crs=ccrs.PlateCarree())
    ax.set_boundary(bound_in_target)
    g=ax.gridlines(color='black',alpha=0.5,linestyle='--',draw_labels=True,x_inline=False, y_inline=False,xlocs=xticks,ylocs=yticks)
    g.top_labels=False; g.right_labels=False; g.rotate_labels=False
    g.xformatter=LongitudeFormatter(direction_label=dir_labels); g.yformatter=LatitudeFormatter(direction_label=dir_labels)
    g.xlabel_style = {'size': 8, 'color': 'gray'}; g.ylabel_style = {'size': 8, 'color': 'gray'}
    return mp,g
    
```
